```python
main = pd.read_csv("PD 2023 Wk 1 Input.csv")
target = pd.read_csv("Targets.csv")
```

```python
main.head()
```

```python
main['Bank'] = main['Transaction Code'].str.split('-',expand = True)[0]
```

```python
main_df = main[main['Bank'].str.contains("DSB")]
```

```python
main_df.head()
```

```python
main_df["Online or In-Person"] = np.where(main_df["Online or In-Person"] == 1 , "Online" , "In-Person")
```

```python
main_df.dtypes
```

```python
main_df["Transaction Date"] = pd.to_datetime(main_df['Transaction Date'], format='%d/%m/%Y %H:%M:%S')
```

```python
main_df.head()
```

```python
main_df["Quarter"] = main_df["Transaction Date"].dt.quarter
```

```python
main_df.head()
```

```python
main_df = main_df.groupby(["Online or In-Person","Quarter"], as_index = False)["Value"].sum()
```

```python
main_df
```

```python
target = pd.melt(target, id_vars=['Online or In-Person'], value_vars=['Q1','Q2','Q3','Q4'])
```

```python
target.head(10)
```

```python
target.columns = ['Online or In-Person','Quarter','Quarterly Targets']
```

```python
target['Quarter'] = target['Quarter'].str.replace('Q','')
target['Quarter'] = target['Quarter'].astype(int)
```

```python
target.head(8)
```

```python
df = main_df.merge(target, on =['Online or In-Person','Quarter'])
```

```python
df.head(8)
```

```python
df['Variance to Targets'] = df['Value'] - df['Quarterly Targets']
```
